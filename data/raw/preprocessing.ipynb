```python
import pandas as pd

# ----------------------------
# Paths — edit these
# ----------------------------
INSTITUTION_FILE = "cohorts_institutions.csv"
FIELD_OF_STUDY_FILE = "recent-cohorts-filed.csv"

# ----------------------------
# Helper to list columns
# ----------------------------
def show_columns(path, n_preview=50):
    print(f"\n=== Loading schema for: {path} ===")

    # read ONLY the header row
    with open(path, "r", encoding="utf-8") as f:
        header = f.readline().strip().split(",")

    print(f"Total columns: {len(header)}")

    # preview
    print(f"\nFirst {n_preview} columns:\n")
    for col in header[:n_preview]:
        print(col)

    print("\n--- End of preview ---")
    return header

# ----------------------------
# Run
# ----------------------------
if __name__ == "__main__":
    inst_cols = show_columns(INSTITUTION_FILE)
    fos_cols = show_columns(FIELD_OF_STUDY_FILE)

    # Save full column lists so you can inspect later
    pd.Series(inst_cols).to_csv("institution_columns.txt", index=False)
    pd.Series(fos_cols).to_csv("field_of_study_columns.txt", index=False)

    print("\nColumn lists saved as:")
    print(" - institution_columns.txt")
    print(" - field_of_study_columns.txt")
```
