```python
# Python
import pandas as pd

df = pd.read_csv("clean_data.csv")

def major_outcome(df, major_substr, credlev=None):
    mask = df["CIPDESC"].str.contains(major_substr, case=False, na=False)
    if credlev is not None:
        mask &= (df["CREDLEV"] == credlev)
    cols = [
        "INSTNM","CIPDESC","CREDLEV",
        "EARN_MDN_1YR","EARN_MDN_4YR","EARN_MDN_5YR",
        "MD_EARN_WNE_P6","MD_EARN_WNE_P10","MD_EARN_WNE_5YR",
        "DEBT_MDN","RPY_1YR_RT","RPY_3YR_RT","RPY_5YR_RT",
        "NPT4_PUB","NPT4_PRIV","COSTT4_A"
    ]
    out = df.loc[mask, cols].copy()
    # Coerce to numeric, ignoring privacy-suppressed strings
    for c in cols[3:]:
        out[c] = pd.to_numeric(out[c], errors="coerce")
    # Aggregate across institutions for a quick view
    summary = out.describe(include="all").T[["count","mean","50%","min","max"]]
    return out.sort_values("MD_EARN_WNE_P10", ascending=False), summary

# Example: outcomes for “Computer and Information Sciences” bachelor’s (CREDLEV 3)
# detail, summary = major_outcome(df, "Computer and Information Sciences", credlev=3)
detail, summary = major_outcome(df, "Biology", credlev=3)
summary
```

### Utilities: majors picker and automatic cred level


```python
import re
import numpy as np
import pandas as pd
from typing import Optional, List, Tuple

# Return unique majors (CIP descriptions) sorted
def list_majors(df: pd.DataFrame) -> List[str]:
    col = "CIPDESC"
    if col not in df.columns:
        raise KeyError(f"Expected column '{col}' not found")
    majors = (
        df[col]
        .dropna()
        .astype(str)
        .str.strip()
        .replace("", np.nan)
        .dropna()
        .unique()
        .tolist()
    )
    return sorted(majors, key=lambda x: x.lower())

# Filter majors by substring or regex (case-insensitive)
def find_majors(df: pd.DataFrame, query: str, use_regex: bool = False, limit: int = 50) -> List[str]:
    majors = list_majors(df)
    if not query:
        return majors[:limit]
    if use_regex:
        pat = re.compile(query, flags=re.IGNORECASE)
        found = [m for m in majors if pat.search(m)]
    else:
        q = query.lower()
        found = [m for m in majors if q in m.lower()]
    return found[:limit]

# Heuristic to pick a default CREDLEV for a subset:
# - If credlev provided -> use it.
# - Else pick the most frequent CREDLEV among matched rows.
def pick_credlev(df: pd.DataFrame, major_substr: str, credlev: Optional[int] = None) -> Optional[int]:
    if credlev is not None:
        return credlev
    mask = df["CIPDESC"].str.contains(major_substr, case=False, na=False)
    if "CREDLEV" not in df.columns or not mask.any():
        return None
    c = df.loc[mask, "CREDLEV"].dropna()
    if c.empty:
        return None
    return int(c.mode().iloc[0])

# Wrapper: choose major and cred level, then run major_outcome
def choose_and_analyze(
    df: pd.DataFrame,
    major_query: str,
    credlev: Optional[int] = None,
    show_candidates: int = 10
) -> Tuple[pd.DataFrame, pd.DataFrame, dict]:
    candidates = find_majors(df, major_query, use_regex=False, limit=show_candidates)
    if not candidates:
        raise ValueError(f"No majors matched query: '{major_query}'")
    chosen_major = candidates[0]  # pick best match (first)
    chosen_credlev = pick_credlev(df, chosen_major, credlev)
    detail, summary = major_outcome(df, chosen_major, credlev=chosen_credlev)
    meta = {
        "query": major_query,
        "chosen_major": chosen_major,
        "chosen_credlev": chosen_credlev,
        "candidates_preview": candidates,
    }
    return detail, summary, meta
```

### Interactive picker: choose major and credential level


```python
import ipywidgets as W
from IPython.display import display, clear_output

# Prepare choices
_all_majors = list_majors(df)

def _majors_filtered(query: str) -> list[str]:
    return find_majors(df, query or "", limit=200)

# Widgets
search_box = W.Text(
    value="biology",
    placeholder="Type to filter majors (substring)...",
    description="Search:",
    disabled=False,
    layout=W.Layout(width="400px")
)

majors_dd = W.Dropdown(
    options=_majors_filtered(search_box.value),
    description="Major:",
    layout=W.Layout(width="600px")
)

credlev_dd = W.Dropdown(
    options=[None, 1, 2, 3, 4, 5, 6, 7],  # adjust if your dataset uses a subset
    value=None,
    description="CREDLEV:",
    layout=W.Layout(width="200px")
)

auto_credlev_btn = W.Button(
    description="Auto-pick CREDLEV",
    tooltip="Pick the most frequent cred level for selected major",
    button_style="info",
)

run_btn = W.Button(
    description="Analyze",
    button_style="primary"
)

out = W.Output()

# Wire up interactions
def on_search_change(change):
    if change["name"] == "value":
        options = _majors_filtered(change["new"])
        majors_dd.options = options
        if options:
            majors_dd.value = options[0]

search_box.observe(on_search_change)

def on_auto_pick_clicked(_):
    if majors_dd.value:
        picked = pick_credlev(df, majors_dd.value, credlev=None)
        credlev_dd.value = picked

auto_credlev_btn.on_click(on_auto_pick_clicked)

def on_analyze_clicked(_):
    with out:
        clear_output()
        chosen_major = majors_dd.value
        chosen_cred = credlev_dd.value
        detail, summary = major_outcome(df, chosen_major, credlev=chosen_cred)
        print(f"Major: {chosen_major}")
        print(f"CREDLEV: {chosen_cred}")
        display(summary.head(12))
        display(detail.head(20))

run_btn.on_click(on_analyze_clicked)

controls = W.HBox([search_box, majors_dd, credlev_dd, auto_credlev_btn, run_btn])
display(W.VBox([controls, out]))
```

### Generate JSON for docs/graph.html picker (majors, credlevs, pre-binned hist + percentiles)


```python
# Produces docs/data.json consumed by docs/graph.html
# Requires a DataFrame `df` with columns: CIPDESC, CREDLEV, EARN_MDN_1YR

import os
import json
import numpy as np
import pandas as pd

if "df" not in globals():
    # Fallback: try to load if not present
    if os.path.exists("clean_data.csv"):
        df = pd.read_csv("clean_data.csv")
    else:
        raise RuntimeError("DataFrame `df` not found and clean_data.csv is missing.")

# Build choices
majors = (
    df["CIPDESC"].dropna().astype(str).str.strip()
      .replace("", np.nan).dropna().unique().tolist()
)
majors = sorted(majors, key=str.lower)

has_cred = "CREDLEV" in df.columns
credlevs: list[int] = []
if has_cred:
    credlevs = sorted({int(c) for c in pd.to_numeric(df["CREDLEV"], errors="coerce").dropna().unique().tolist()})

def slice_series(major: str, cred: int | None) -> pd.Series:
    m = df["CIPDESC"].astype(str).str.contains(major, case=False, na=False)
    if has_cred and cred is not None:
        m &= (pd.to_numeric(df["CREDLEV"], errors="coerce") == cred)
    return pd.to_numeric(df.loc[m, "EARN_MDN_1YR"], errors="coerce").dropna()

def to_payload(s: pd.Series, bins: int = 50) -> dict:
    if s.empty:
        return {"bins": [], "pcts": {}}
    counts, edges = np.histogram(s.values, bins=bins)
    bins_list = [
        {"x0": float(edges[i]), "x1": float(edges[i+1]), "count": int(counts[i])}
        for i in range(len(counts))
    ]
    pcts = {f"p{k}": float(np.percentile(s.values, k)) for k in (10, 25, 50, 75, 90)}
    return {"bins": bins_list, "pcts": pcts}

data = {"majors": majors, "credlevs": credlevs, "payload": {}}
for m in majors:
    data["payload"][m] = {}
    keys = credlevs if has_cred else [None]
    for c in keys:
        s = slice_series(m, c)
        data["payload"][m][(None if not has_cred else int(c))] = to_payload(s)

os.makedirs("docs", exist_ok=True)
out_path = os.path.join("docs", "data.json")
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(data, f)

print(f"Wrote {out_path} with {len(majors)} majors and {len(credlevs)} cred levels.")
# ... existing code ...
```
